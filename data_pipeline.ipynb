```python
import pandas as pd
import os
```

```python
#define file path
file_path = "C:\\Users\\vaidy\\Downloads\\dynamic_pricing.csv"
```

```python
#create ingestion function
def load_data(file_path):
    try:
        data = pd.read_csv(file_path)
        print("Data successfully loaded.")
        print(f"Rows: {data.shape[0]}, Columns: {data.shape[1]}")
        return data
    except FileNotFoundError:
        print(" File not found. Check your path.")
        return None
```

```python
#clean data
def clean_data(data):
    # remove duplicates
    data = data.drop_duplicates()
    
    # handle missing values
    data = data.dropna()
    
    # rename columns (optional)
    data.columns = [col.strip().replace(" ", "_") for col in data.columns]
    
    print("Data cleaning completed.")
    return data
```

```python
def save_data(data, output_path="C:/Users/vaidy/OneDrive/Desktop/clean_data.csv"):
    # Create folder if it doesn't exist
    folder = os.path.dirname(output_path)
    if folder and not os.path.exists(folder):
        os.makedirs(folder)
        print(f"ðŸ“‚ Folder created at {folder}")

    # Save DataFrame to CSV
    data.to_csv(output_path, index=False)
    print(f"âœ… Clean data saved to {output_path}")

# Example usage (pass your cleaned DataFrame here)
# save_data(data)
```

```python
#run the pipeline
def data_ingestion_pipeline(file_path):
    data = load_data(file_path)
    if data is not None:
        data = clean_data(data)
        save_data(data)
        return data
file_path = "C:\\Users\\vaidy\\Downloads\\titanic.csv"
# Run the pipeline
df = data_ingestion_pipeline(file_path)

print(df.head())
```
